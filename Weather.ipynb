```python
def get_index(location, date, time):

    df = weather.where((weather['hour_datetime']==time)&(weather['city_resolvedAddress']==location)&(weather['day_datetime']==date))
    
    p = df.loc[(weather['hour_datetime']==time)&(weather['city_resolvedAddress']==location)&(weather['day_datetime']==date)].index
    
    return p
```
